```python
import json
import requests
  
# Defining Binance API URL

key = "https://api.binance.com/api/v3/ticker/price?symbol="
  
# Making list for multiple crypto's

currencies = list_symbols
k = 0

while k <100:

    j = 0
    
    # running loop to print all crypto prices
    for i in currencies:
        # completing API for request
        url = key+currencies[j]  
        data = requests.get(url)
        data = data.json()
        if j == 0:
            BTCUSDT.append(data['price'])
        elif j == 1:
            ETHUSDT.append(data['price'])
        elif j == 2:
            BNBUSDT.append(data['price'])
        elif j == 3:
            BCCUSDT.append(data['price'])
        elif j == 4:
            NEOUSDT.append(data['price'])
        elif j == 5:
            LTCUSDT.append(data['price'])
        elif j == 6:
            QTUMUSDT.append(data['price'])
        elif j == 7:
            ADAUSDT.append(data['price'])
        elif j == 8:
            XRPUSDT.append(data['price'])
        elif j == 9:
            EOSUSDT.append(data['price'])
        elif j == 10:
            TUSDUSDT.append(data['price'])
        elif j == 11:
            IOTAUSDT.append(data['price'])
        elif j == 12:
            XLMUSDT.append(data['price'])
        elif j == 13:
            ONTUSDT.append(data['price'])
        elif j == 14:
            TRXUSDT.append(data['price'])
        elif j == 15:
            ETCUSDT.append(data['price'])
        elif j == 16:
            ICXUSDT.append(data['price'])
        elif j == 17:
            VENUSDT.append(data['price'])
        elif j == 18:
            NULSUSDT.append(data['price'])
        elif j == 19:
            VETUSDT.append(data['price'])

        
        j += 1
        k += 1
        
        print(f"{data['symbol']} price is {data['price']}")
    time.sleep(15)
```
